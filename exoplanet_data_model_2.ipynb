# Read the CSV and Perform Basic Data Cleaning

```python
df = pd.read_csv("exoplanet_data.csv")
# Drop the null columns where all values are null
df = df.dropna(axis='columns', how='all')
# Drop the null rows
df = df.dropna()
df.head(100)
```

```python
df_drop_disposition=df.drop(['koi_disposition'], axis=1)
df_drop_disposition
```

```python
koi_disposition=df['koi_disposition']

data = koi_disposition.copy()
```

# Select your features (columns)

```python
# define dataset

X = df_drop_disposition
y= koi_disposition

```

# Create a Train Test Split

Use `koi_disposition` for the y values

```python
#create train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)


label_encoder=LabelEncoder()
label_encoder.fit(y_train)
y_train_encoded=label_encoder.transform(y_train)
y_test_encoded=label_encoder.transform(y_test)

y_train_categorical=to_categorical(y_train_encoded)
y_test_categorical=to_categorical(y_test_encoded)

print(y_test_categorical)
```

```python
X_train.head()
```

# Pre-processing

Scale the data using the MinMaxScaler and perform some feature selection

```python
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.layers import Dense

# Create a StandardScater model and fit it to the training data
X_scaler = StandardScaler().fit(X_train)

# Scale your data
X_train_scaled=X_scaler.transform(X_train)
X_test_scaled=X_scaler.transform(X_test)

from tensorflow import keras
from tensorflow.keras import layers

# Define Sequential model with 3 layers
model = keras.Sequential(
    [
        layers.Dense(50, activation="relu", name="layer1", input_dim=40),
        layers.Dense(25, activation="relu", name="layer2" ),
        layers.Dense(3, activation= "softmax", name="layer3"),
    ]
)


model.summary()
```

```python
opt = keras.optimizers.Adam(learning_rate=0.01)
model.compile(loss='categorical_crossentropy', optimizer=opt, metrics=['accuracy'])
```

```python
print(len(X_train_scaled))
print(len(y_train_categorical))
```

```python
# # fit the model
# model.fit(X_train_scaled, y_train_categorical)
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
#test the model
model.evaluate(X_test_scaled, y_test_categorical)
```

# Train the Model



```python
print(f"Training Data Score: {model.evaluate(X_train_scaled, y_train_categorical)}")
print(f"Testing Data Score: {model.evaluate(X_test_scaled, y_test_categorical)}")
```
